```python
import os
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
dir_dataset = "/home/abhishek/Desktop/RUG/ai_master_thesis/kinect_data/transparent_objects/glass_tumbler_2/"
```

```python
file_depth = os.path.join(dir_dataset, "depth_images.pickle")
file_color = os.path.join(dir_dataset, "color_images.pickle")
```

```python
def load_all_images(file_pickle):
    with open(file_pickle, "rb") as file_handler:
        while True:
            try:
                yield pickle.load(file_handler)
            except EOFError:
                break
```

```python
color_images = list(load_all_images(file_color))
```

```python
depth_images = list(load_all_images(file_depth))
```

```python
color_images = np.array(color_images)
depth_images = np.array(depth_images)
```

```python
print(color_images.shape, color_images.dtype)
print(depth_images.shape, depth_images.dtype)
```

```python
image_index = 5
```

```python
plt.figure()
plt.title("Color image")
plt.imshow(color_images[image_index])
```

```python
plt.figure()
plt.title("Depth image")
plt.imshow(depth_images[image_index], cmap="gray")
```
